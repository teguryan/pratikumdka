A.Graf Peta peta besar di Eropa

```python
import networkx as nx
import matplotlib.pyplot as plt

# Buat graph
G = nx.Graph()

# Tambahkan edges dan jaraknya (sesuai gambar)
edges = [
    ("Oradea", "Zerind", 71),
    ("Zerind", "Arad", 75),
    ("Arad", "Sibiu", 140),
    ("Arad", "Timisoara", 118),
    ("Timisoara", "Lugoj", 111),
    ("Lugoj", "Mehadia", 70),
    ("Mehadia", "Drobeta", 75),
    ("Drobeta", "Craiova", 120),
    ("Craiova", "Rimnicu Vilcea", 146),
    ("Craiova", "Pitesti", 138),
    ("Rimnicu Vilcea", "Sibiu", 80),
    ("Rimnicu Vilcea", "Pitesti", 97),
    ("Sibiu", "Fagaras", 99),
    ("Fagaras", "Bucharest", 211),
    ("Pitesti", "Bucharest", 101),
    ("Bucharest", "Giurgiu", 90),
    ("Bucharest", "Urziceni", 85),
    ("Urziceni", "Hirsova", 98),
    ("Hirsova", "Eforie", 86),
    ("Urziceni", "Vaslui", 142),
    ("Vaslui", "Iasi", 92),
    ("Iasi", "Neamt", 87),
    ("Sibiu", "Oradea", 151)
]

for u, v, w in edges:
    G.add_edge(u, v, weight=w)


pos = {
    "Oradea": (1, 8),
    "Zerind": (2, 7),
    "Arad": (3, 6),
    "Timisoara": (2, 5),
    "Lugoj": (3, 4.2),
    "Mehadia": (4, 3.5),
    "Drobeta": (3, 2.5),
    "Craiova": (4.5, 2.5),
    "Rimnicu Vilcea": (5, 4.5),
    "Pitesti": (6, 3.2),
    "Sibiu": (4.5, 6),
    "Fagaras": (6, 6.5),
    "Bucharest": (7, 2.5),
    "Giurgiu": (7, 1.5),
    "Urziceni": (8, 3.5),
    "Hirsova": (9, 3.5),
    "Eforie": (9.5, 2.5),
    "Vaslui": (9, 5),
    "Iasi": (9, 6),
    "Neamt": (9, 7),
}

# Depth-First Search (DFS) dari Arad
dfs_edges = list(nx.dfs_edges(G, source="Arad"))
dfs_nodes = [node for node, _ in dfs_edges]
print(f"Nodes yang dikunjungi dimulai dari Arad: {dfs_nodes}")

# Gambar graf
plt.figure(figsize=(14, 10))
nx.draw(G, pos, with_labels=True, node_size=2500, node_color="lightgreen", font_size=9, font_weight="bold")
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='blue', font_size=8)

# Gambar jalur DFS
dfs_edges_set = set(dfs_edges)
edges_to_highlight = [(u, v) for u, v in G.edges() if (u, v) in dfs_edges_set or (v, u) in dfs_edges_set]
nx.draw_networkx_edges(G, pos, edgelist=edges_to_highlight, edge_color="red", width=2)

plt.title("Graf Peta Kota-Kota Besar di Eropa dengan DFS dari Arad", fontsize=14)
plt.axis('off')
plt.show()
```

1. networkx digunakan untuk membuat dan mengelola graf.
2. matplotlib.pyplot digunakan untuk menggambar graf secara visual.
3. G adalah objek graf tak berarah (undirected graph). Kalau kamu ingin membuat graf berarah, gunakan nx.DiGraph().
4. Daftar edges menyimpan informasi hubungan antar kota dan jaraknya (bobot).
5. G.add_edge(u, v, weight=w) menambahkan edge dari u ke v dengan bobot w.
6. pos adalah dictionary yang menentukan posisi setiap node pada graf agar tampak seperti peta.
7. nx.dfs_edges(G, source="Arad") melakukan penelusuran DFS mulai dari kota Arad.
8. Hasilnya adalah daftar edge yang dilalui selama DFS.
9. dfs_nodes menyimpan urutan node yang dikunjungi.
10. nx.draw() menggambar node dan edge-nya.
11. nx.draw_networkx_edge_labels() menampilkan bobot edge (jarak antar kota).
12. plt.axis('off') menyembunyikan sumbu koordinat agar terlihat seperti peta.

B. Graf Peta peta besar di Indonesia (Pulau Jawa)

```python
import networkx as nx
import matplotlib.pyplot as plt

# Inisialisasi graf
G = nx.Graph()

# Menambahkan edge dan bobot jarak antar kota
edges = [
    ('Jakarta', 'Cirebon', 327),
    ('Jakarta', 'Bandung', 270),
    ('Cirebon', 'Bandung', 120),
    ('Cirebon', 'Yogyakarta', 210),
    ('Cirebon', 'Semarang', 305),
    ('Bandung', 'Yogyakarta', 373),
    ('Semarang', 'Yogyakarta', 109),
    ('Semarang', 'Surakarta', 97),
    ('Semarang', 'Surabaya', 369),
    ('Surabaya', 'Malang', 94),
    ('Surakarta', 'Malang', 370),
    ('Surakarta', 'Yogyakarta', 60),
]
G.add_weighted_edges_from(edges)

# Posisi manual sesuai layout dari peta
pos = {
    'Jakarta': (0, 0),
    'Bandung': (-1, 1),
    'Cirebon': (-2, 2),
    'Semarang': (1, 1),
    'Yogyakarta': (-1, -1),
    'Surakarta': (-1, -2),
    'Surabaya': (-1, -3),
    'Malang': (-1, -4),
}

# Fungsi DFS
def dfs(graph, start, goal, path=None, visited=None):
    if visited is None:
        visited = set()
    if path is None:
        path = [start]

    visited.add(start)

    if start == goal:
        return path

    for neighbor in graph.neighbors(start):
        if neighbor not in visited:
            result = dfs(graph, neighbor, goal, path + [neighbor], visited)
            if result:
                return result

    return None  # Tidak ditemukan

# Jalankan DFS
start_city = "Bandung"
goal_city = "Surakarta"
path_result = dfs(G, start_city, goal_city)

# Gambar graf
plt.figure(figsize=(12, 8))
nx.draw(
    G, pos,
    with_labels=True,
    node_color='red',
    edge_color='black',
    node_size=1000,
    font_color='white',
    font_size=6,
    font_weight='bold',
    width=2
)

# Tambahkan label bobot di atas edge
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color='green',
    font_size=9,
    label_pos=0.6
)

# Highlight jalur hasil DFS
if path_result:
    edge_path = list(zip(path_result, path_result[1:]))  # Buat pasangan edge dari path
    nx.draw_networkx_edges(G, pos, edgelist=edge_path, edge_color='blue', width=4)
    print(f"Jalur DFS dari {start_city} ke {goal_city}:")
    print(" -> ".join(path_result))
else:
    print("Jalur tidak ditemukan.")

plt.title("Graf Kota (Jalur DFS Berwarna Biru)", fontsize=14)
plt.axis("off")
plt.show()
```

1. networkx: untuk membuat dan mengelola graf.
2. matplotlib.pyplot: untuk menggambar graf secara visual.
3. G = nx.Graph() Membuat graf tak berarah (Graph()), cocok untuk jaringan kota seperti ini.
4. Daftar edges berisi pasangan kota dan jarak antar kota (sebagai bobot).
5. add_weighted_edges_from menambahkan edge dengan format (node1, node2, weight).
6. pos menentukan koordinat setiap kota agar tampilan graf menyerupai layout geografis pulau Jawa secara sederhana.
7. visited digunakan untuk menyimpan node yang sudah dikunjungi agar tidak loop.
8. path menyimpan jalur dari start ke goal.
9. Jika path_result ditemukan, maka jalur DFS di-highlight dengan warna biru.
10. zip(path_result, path_result[1:]) dipakai untuk membuat pasangan edge dari urutan node.
